### Cleaning Data

```python
# Check for issues
df.isnull().sum()       # null count per column
df.duplicated().sum()   # duplicate row count
df.dtypes               # column types
```

```python
# Handling Nulls
df.dropna()                        # drop rows with any null
df.dropna(subset=['amount'])       # drop only if 'amount' is null
df.fillna(0)                       # fill all nulls with 0
df['amount'].fillna(df['amount'].mean())  # fill with mean
```

```python
# removing duplicates
df.drop_duplicates()                      # remove duplicate rows (entire row has to match)
df.drop_duplicates(subset=['order_id'])   # based on the column (column value has to match)
```

```python
# Fixing Data types
df['amount'] = df['amount'].astype(float)
df['date'] = pd.to_datetime(df['date'])
df['order_id'] = df['order_id'].astype(str)
```

```python
# Cleaning text
df['region'] = df['region'].str.strip()    # remove whitespace
df['region'] = df['region'].str.lower()    # lowercase
df['region'] = df['region'].str.replace('$', '', regex=False)
```
